```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
normie = np.random.normal(10, 2, 100)
plt.hist(normie, bins=20, color='orange')
plt.axvline(normie.mean(), color='green', linestyle='solid', linewidth=3)
plt.axvline(normie.mean()+normie.std(), color='blue', linestyle='dashed', linewidth=5)
plt.axvline(normie.mean()-normie.std(), color='blue', linestyle='dashed', linewidth=5)
plt.title('Normal Distribution')
plt.show();
print(format("Mean:"), normie.mean())
print(format("Standard Deviation:"), normie.std())
```

The mean is in the center of the distribution with one standard deviation above and below the mean containing 67% of the data. This Normal Distribution is a measure of central tendency.

```python
poison = np.random.poisson(3, 100)
plt.hist(poison, bins=10, color='purple')
plt.axvline(poison.mean(), color='green', linestyle='solid', linewidth=5)
plt.axvline(poison.mean()+poison.std(), color='black', linestyle='dashed', linewidth=5)
plt.axvline(poison.mean()-poison.std(), color='black', linestyle='dashed', linewidth=5)
plt.title('Poisson Distribution')
plt.show();
print(format("Mean:"), poison.mean())
print(format("Standard Deviation:"), poison.std())
```

With the majority of the data below the mean, the poisson distribution is not a measure of central tendency because it is skewed to the right.

```python
binome = np.random.binomial(10, 0.2, 100)
plt.hist(binome, bins=10, color='blue')
plt.axvline(binome.mean(), color='gold', linestyle='solid', linewidth=2)
plt.axvline(binome.mean()+binome.std(), color='red', linestyle='dashed', linewidth=2)
plt.axvline(binome.mean()-binome.std(), color='red', linestyle='dashed', linewidth=2)
plt.title('Binomial Distribution')
plt.show();
print(format("Mean:"), binome.mean())
print(format("Standard Deviation:"), binome.std())
```

Binomial distribution can be a measure of central tendency, but in this case the graph is skewed to the right.

```python
gammy = np.random.gamma(3, 0.2, 100)
plt.hist(gammy, bins=15, color='orange')
plt.axvline(gammy.mean(), color='purple', linestyle='solid', linewidth=2)
plt.axvline(gammy.mean()+gammy.std(), color='blue', linestyle='dashed', linewidth=2)
plt.axvline(gammy.mean()-gammy.std(), color='blue', linestyle='dashed', linewidth=2)
plt.title('Gamma Distribution')
plt.show();
print(format("Mean:"), gammy.mean())
print(format("Standard Deviation:"), gammy.std())
```

The majority of the data is below the mean, so the gamma distribution is not a measure of central tendency because it is skewed to the right.

```python
loganormie = np.random.lognormal(1, 1, 100)
plt.hist(loganormie, bins=5, color='purple')
plt.axvline(loganormie.mean(), color='black', linestyle='solid', linewidth=2)
plt.axvline(loganormie.mean()+loganormie.std(), color='green', linestyle='dashed', linewidth=2)
plt.axvline(loganormie.mean()-loganormie.std(), color='green', linestyle='dashed', linewidth=2)
plt.title("Log-normal Distribution")
plt.show();
print(format("Mean:"), loganormie.mean())
print(format("Standard Deviation:"), loganormie.std())
```

As you can see the data is skewed to the right with barely any data below one standard deviation under the mean.

```python
multinoms = np.random.multinomial(100, [2/8]*4, size=100)
plt.hist(multinoms, bins=5)
plt.axvline(multinoms.mean(), color='black', linestyle='solid', linewidth=3)
plt.axvline(multinoms.mean()+multinoms.std(), color='violet', linestyle='dashed', linewidth=3)
plt.axvline(multinoms.mean()-multinoms.std(), color='violet', linestyle='dashed', linewidth=3)
plt.title("Multinomial Distribution")
plt.show();
print(format("Mean:"), multinoms.mean())
print(format("Standard Deviation:"), multinoms.std())
multinoms
```

```python
normie1 = np.random.normal(5, 0.5, 100)
normie2 = np.random.normal(10, 1, 100)
abnormal = normie1 + normie2
plt.hist(abnormal, bins=20, color='red')
plt.axvline(abnormal.mean(), color='black', linestyle='solid', linewidth=5)
plt.axvline(abnormal.mean()+abnormal.std(), color='blue', linestyle='dashed', linewidth=5)
plt.axvline(abnormal.mean()-abnormal.std(), color='blue', linestyle='dashed', linewidth=5)
plt.title("Abnormal Distribution")
plt.show();
print(format("Mean:"), abnormal.mean())
print(format("Standard Deviation:"), abnormal.std())
```

Two normally distributed variables have developed a great relationship based on the datapoints, even with a difference of mean and standard deviation. With a different distribution added to the normal distribution, it is likely that the relationship with the data points may not have measures of central tendency.
